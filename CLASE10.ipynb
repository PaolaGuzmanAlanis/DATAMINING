```python
ndf = ndf.drop('nan', axis = 1)
ndf.columns
```

## FP frowth algorithm

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
freq_itms = fpgrowth(ndf,
                     min_support = 0.6,
                     use_colnames = True)
freq_itms
```

```python
as_rules = association_rules(freq_itms,
                             metric = 'lift',
                             min_threshold = 1.5)
as_rules
```

CASO 1: Which are the top 10 frequent patterns with high confidence?

```python
SortConf = as_rules.sort_values('confidence',
                                  ascending = False).head(5)
SortConf
```
